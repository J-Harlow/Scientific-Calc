```python
import math

#Calculator Functions
def factorial(n):
    if n==0:
        return 1
    else:
        return n*factorial(n - 1)
def square_root(x):
    return math.sqrt(x)
def power(x, y):
    return x**y
def logarithm(x, base):
    return math.log(x, base)
def sine(x):
    return math.sin(math.radians(x))
def cosine(x):
    return math.cos(math.radians(x))
def tangent(x):
    return math.tan(math.radians(x))
def inverse_sine(x):
    return math.degrees(math.asin(x))
def inverse_cosine(x):
    return math.degrees(math.acos(x))
def inverse_tangent(x):
    return math.degrees(math.atan(x))
def hyperbolic_sine(x):
    return math.sinh(x)
def hyperbolic_cosine(x):
    return math.cosh(x)
def hyperbolic_tangent(x):
    return math.tanh(x)

#Interface
print("Available Operations:")
print("1. Factorial")
print("2. Square Root")
print("3. Power")
print("4. Logarithm")
print("5. Sine")
print("6. Cosine")
print("7. Tangent")
print("8. Inverse Sine")
print("9. Inverse Cosine")
print("10. Inverse Tangent")
print("11. Hyperbolic Sine")
print("12. Hyperbolic Cosine")
print("13. Hyperbolic Tangent")

choice=int(input("Enter the operation number: "))

if choice in range(1, 14):
    if choice in (1, 2, 5, 6, 7, 11, 12, 13):
        x=float(input("Enter the value: "))
    else:
        x=float(input("Enter the first value: "))
        y=float(input("Enter the second value: "))

    if choice==1:
        result=factorial(x)
    elif choice==2:
        result=square_root(x)
    elif choice==3:
        result=power(x, y)
    elif choice==4:
        base=float(input("Enter the logarithm base: "))
        result=logarithm(x, base)
    elif choice==5:
        result=sine(x)
    elif choice==6:
        result=cosine(x)
    elif choice==7:
        result=tangent(x)
    elif choice==8:
        result=inverse_sine(x)
    elif choice==9:
        result=inverse_cosine(x)
    elif choice==10:
        result=inverse_tangent(x)
    elif choice==11:
        result=hyperbolic_sine(x)
    elif choice==12:
        result=hyperbolic_cosine(x)
    elif choice==13:
        result=hyperbolic_tangent(x)

    print("Result: ", result)

else:
    print("Invalid choice. Please select a valid operation.")
```
